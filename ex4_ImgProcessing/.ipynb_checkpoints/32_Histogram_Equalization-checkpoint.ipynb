```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load gorilla as grayscale
gorilla=cv2.imread('../DATA/gorilla.jpg',0)
```

```python
def display(img, cmap=None):
    fig=plt.figure(figsize=(10,8))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
display(gorilla)
```
