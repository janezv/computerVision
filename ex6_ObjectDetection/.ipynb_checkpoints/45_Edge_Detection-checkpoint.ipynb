```python
import cv2      
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('../DATA/sammy_face.jpg')
cv2.imshow('img',img)
```

```python
plt.imshow(img)
```

```python
edges=cv2.Canny(image=img, threshold1=127,threshold2=128)
plt.imshow(edges)
```

```python
edges=cv2.Canny(image=img, threshold1=0,threshold2=255)
plt.imshow(edges)
```

```python
# calculate the median value of the image
med_val=np.median(img)
med_val
```

```python
#set lower value to either 0 pr 70% of the median value whichever is greater
lower=int(max(0,0.7*med_val))
#set upper value to either 255 or 130% of the median value whichever is smaller
upper=int(min(255,1.3*med_val))
```

```python
edges=cv2.Canny(image=img, threshold1=lower,threshold2=upper+100)
plt.imshow(edges)
```

### Predhodno zabluraj img

```python
blurrre_img=cv2.blur(img,ksize=(5,5))
edges=cv2.Canny(image=blurrre_img, threshold1=lower,threshold2=upper)
plt.imshow(edges)
```
