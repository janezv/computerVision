```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

### Branje img in pretovrba iz bgr v rgb. plt.imsho() v jupyter-ju ne dela brez te prevorbe

```python
img=cv2.imread("00-puppy.jpg")
type(img)
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

### Za prilagajanje slike v jupyter-ju


```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
ax.imshow(fix_img)
```

```python
fig = plt.figure(figsize=(2, 2))
ax = fig.add_subplot(111)
ax.imshow(fix_img)
```
